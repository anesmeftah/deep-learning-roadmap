```python
for col in ['Quantity', 'Price Per Unit', 'Total Spent', 'Item']:
    data[col] = data[col].replace(['ERROR', 'UNKNOWN'], np.nan)
```

```python
data['Quantity'].unique()
```

```python
data['Quantity'] = data['Quantity'].astype(float).copy()
```

```python
data['Quantity'].unique()
```

```python
data['Price Per Unit'] = data['Price Per Unit'].astype(float).copy()
data['Total Spent'] = data['Total Spent'].astype(float).copy()
```

```python
data['Item'].unique()
# Coffee
```

```python
items = ['Coffee', 'Cake', 'Cookie', 'Salad', 'Smoothie',
       'Sandwich', 'Juice', 'Tea']
for i in items:
    subset = data[data['Item'] == i]
    if not subset.empty:
        print(f"{i}: {subset.iloc[0]['Price Per Unit']}")
    else:
        print(f"{i}: Not found in dataset")
```

```python
#Fill missing prices
for i in items:
    subset = data[data['Item'] == i]
    if not subset.empty:
        price = subset.iloc[0]['Price Per Unit']
        data.loc[data['Item'] == i, 'Price Per Unit'] = price
    else:
        print(f"{i}: Not found in dataset")
```

```python
# Fill missing items
prices = [1.0 , 1.5 , 2 ,5]
for i in prices:
    subset = data[data['Price Per Unit'] == i]
    if not subset.empty:
        item = subset.iloc[0]['Item']
        data.loc[data['Price Per Unit'] == i, 'Item'] = item
    else:
        print(f"{i}: Not found in dataset")
```

```python
data.info()
```

```python

data.loc[data['Price Per Unit'].isna() , 'Price Per Unit'] = data.loc[data['Price Per Unit'].isna() , 'Total Spent'] / data.loc[data['Price Per Unit'].isna() , 'Quantity']
```

```python
data.loc[data['Total Spent'].isna() , 'Total Spent'] = data.loc[data['Total Spent'].isna() , 'Quantity'] * data.loc[data['Total Spent'].isna() , 'Price Per Unit']
```
